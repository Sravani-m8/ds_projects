```python
user_movie_matrix = df.pivot_table(index='userId', columns='title', values='rating')
user_movie_matrix.head()
```

```python
user_movie_matrix = user_movie_matrix.fillna(0)
```

```python
user_movie_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_movie_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)
user_similarity_df.head()
```

```python
def get_popular_movies(n=10):
  popular_movies = df.groupby('title').agg({'rating':'mean', 'userId':'count'}).reset_index()
  popular_movies.columns = ['title', 'average_rating', 'number_of_ratings']
  popular_movies = popular_movies.sort_values(by=['number_of_ratings', 'average_rating'], ascending=False)
  return popular_movies['title'].head(n).tolist()

def recommend_movies(user_id, num_recommendations=5):
  if user_id not in user_movie_matrix.index:
    popular_movie_titles = get_popular_movies(num_recommendations)
    return "User not found. Recommending popular movies:\n" + "\n".join(popular_movie_titles)
  else:
    user_index = user_movie_matrix.index.get_loc(user_id)
    sim_scores = user_similarity_df.iloc[user_index]
    similar_users = sim_scores.nlargest(num_recommendations+1).index[1:]
    similar_users_ratings = user_movie_matrix.loc[similar_users]
    mean_ratings = similar_users_ratings.mean(axis=0)
    user_rated_movies = user_movie_matrix.loc[user_id]
    mean_ratings = mean_ratings[user_rated_movies == 0]
    recommendations = mean_ratings.nlargest(num_recommendations)
    return recommendations.index.tolist()

```

```python
user_id = int(input("Enter the user ID: "))
recommended_movies = recommend_movies(user_id, num_recommendations=8)
print("Recommended Movies: ", recommended_movies)
```

# **Recommendations Using SVD**

```python
movies_df = pd.read_csv('/content/movies_df.csv')
movies_df.head()
```

```python
print(movies_df['rating'].min())
print(movies_df['rating'].max())
```

```python
from surprise import Reader, Dataset, SVD, accuracy
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(0.5, 5))
data = Dataset.load_from_df(movies_df[['userId', 'movieId', 'rating']], reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

model = SVD()
model.fit(trainset)
```

```python
predictions = model.test(testset)
rmse = accuracy.rmse(predictions)
```

```python
from surprise.model_selection import cross_validate

cross_validate_results = cross_validate(model, data, measures=['RMSE'], cv=5, verbose=True)
```

```python
cross_val_df = pd.DataFrame(cross_validate_results)
cross_val_df
```

```python
def recommend_movies_svd(user_id, num_recommendations=5):
  all_movie_ids = movies_df['movieId'].unique()
  rated_movies = movies_df[movies_df['userId'] == user_id]['movieId'].values
  movies_to_predict = np.setdiff1d(all_movie_ids, rated_movies)
  predictions = []
  for movie_id in movies_to_predict:
    pred_rating = model.predict(user_id, movie_id).est
    predictions.append((movie_id, pred_rating))

  predictions.sort(key = lambda x: x[1], reverse = True)
  recommended_movie_ids = [pred[0] for pred in predictions[:num_recommendations]]
  recommended_movies = movies_df[movies_df['movieId'].isin(recommended_movie_ids)][['movieId', 'title']].drop_duplicates()
  return recommended_movies['title'].tolist()
```

```python
user_id = int(input("Enter the user ID: "))
recommended_movies = recommend_movies_svd(user_id, num_recommendations=8)
print("Recommended Movies: ", recommended_movies)
```
