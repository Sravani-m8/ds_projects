```python
import ast

def convert(text):
  L = []
  for i in ast.literal_eval(text):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
def convert3(text):
  L = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter < 3:
      L.append(i['name'])
      counter += 1
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies.loc[8, 'crew']
```

```python
def fetch_director(text):
  L = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      L.append(i['name'])
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def clean_space(L):
  L1 = []
  for i in L:
    L1.append(i.replace(" ", ""))
  return L1
```

```python
movies['genres'] = movies['genres'].apply(clean_space)
movies['keywords'] = movies['keywords'].apply(clean_space)
movies['cast'] = movies['cast'].apply(clean_space)
movies['crew'] = movies['crew'].apply(clean_space)
movies.head()
```

```python
movies['genres'] = [' '.join(genre_list) for genre_list in movies['genres']]
movies['overview'] = [' '.join(overview_list) for overview_list in movies['overview']]
movies['cast'] = [' '.join(cast_list) for cast_list in movies['cast']]
movies['crew'] = [' '.join(crew_list) for crew_list in movies['crew']]
movies['keywords'] = [' '.join(keyword_list) for keyword_list in movies['keywords']]

movies['tags'] = movies['genres'] + ' ' + movies['overview'] + ' ' + movies['cast'] + ' ' + movies['crew'] + ' ' + movies['keywords']
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
movies.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
  text = text.translate(str.maketrans('', '', string.punctuation))
  text = text.split()
  text = [lemmatizer.lemmatize(word) for word in text if word not in stop_words]
  text = " ".join(text)
  return text
```

```python
movies['tags'] = movies['tags'].apply(clean_text)
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors, vectors)
```

```python
def content_based_recommender(title, num_recommendations=10):
  title = title.lower()

  if title not in movies['title'].str.lower().values:
    similar_titles = movies[movies['title'].str.lower().str.contains(title)]['title']
    if not similar_titles.empty:
      return "Did you mean one of these titles?\n" + "\n".join(similar_titles)
    else:
      return "No similar titles found. Please check the spelling."

  idx = movies[movies['title'].str.lower() == title].index[0]
  sim_scores = list(enumerate(similarity[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:num_recommendations+1]
  movie_indices = [i[0] for i in sim_scores]
  return movies.iloc[movie_indices]['title']


```

```python
title = input("enter the movie name: ")
recommendations = content_based_recommender(title)
print(recommendations)
```
